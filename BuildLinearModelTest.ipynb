```python
import numpy as np
import pandas as pd
from BuildLinearData import linear_data
from BuildLinearModel import validate_model
```

### Testing the function on one data set

```python
df1 = pd.read_csv("OrderBook_10.csv")
df2 = pd.read_csv("KlineData_10.csv")
data1 = linear_data(df1, df2)
```

```python
df = data1.copy()
train, test = np.split(df, [int(0.8*len(df))])
```

```python
validate_model(train, test, l=5)
```

We see an r2 score of 0 and an accuracy of only 11 percent. However, I feel like this is due to a really small sample space. If the accuracy uncrease with the size of the datasset that will be a good indication that we are on the right track.

We concatenate all the frames together into one.

```python
frames = [data1, data2, data3, data4, data5]

result = pd.concat(frames)
result
```

And now we send it through our model.

```python
df = result.copy()
train, test = np.split(df, [int(0.8*len(df))])
```

```python
validate_model(train, test, l=5)
```

Bam! We see that our accuracy more than doubled while the errors halved which means our model is workable but we need enough data for that.
